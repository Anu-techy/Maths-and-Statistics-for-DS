```python
# creating a sampling distribution of the sampling means
for city,subdf in df.groupby("City"):
    print(city)
    print(subdf.shape)
    
# from each city a sample of 50 bills are taken     
```

```python
sample_means = []

for city,subdf in df.groupby("City"):
    print(city,"     ",subdf["Bill"].mean())
    sample_means.append(subdf["Bill"].mean())
```

```python
sample_means
```

```python
len(sample_means)
```

```python
sns.histplot(sample_means,kde=True)
```

```python
# This looks like a normal distribution, As the sample size increases, the distribution comes to more normal
```
