```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heights_bellcurve.csv")
df.head()
```

```python
df.shape
```

```python
sns.histplot(df["height"],kde = True)
```

```python
avg = round(df["height"].mean(),2)
std = round(df["height"].std(),2)
```

```python
avg,std
```

```python
sns.histplot(df["height"],kde = True)
plt.axvline(avg, linestyle = "dashed")
#plt.axvline(std = avg, linestyle = "dashed")
```

```python
sns.histplot(df["height"],kde = True)
plt.axvline(avg, color = "black", linestyle = "dashed")
plt.axvline(avg+std, color = "red", linestyle = "dashed")
plt.axvline(avg-std, color = "red", linestyle = "dashed")
```

```python
sns.histplot(df["height"],kde = True)
plt.axvline(avg, color = "black", linestyle = "dashed")

plt.axvline(avg+std, color = "r", linestyle = "dashed")
plt.axvline(avg-std, color = "r", linestyle = "dashed")

plt.axvline(avg+2*std, color = "g", linestyle = "dashed")
plt.axvline(avg-2*std, color = "g", linestyle = "dashed")

plt.axvline(avg+3*std, color = "b", linestyle = "dashed")
plt.axvline(avg-3*std, color = "b", linestyle = "dashed")
```

```python
left1 = round(avg-std,2)
right1 = round(avg+std,2)
left1, right1
```

```python
right2 = round(avg+2*std,2)
left2 = round(avg-2*std,2)
left2, right2
```

```python
right3 = round(avg+3*std,2)
left3 = round(avg-3*std,2)
left3, right3
```

```python
df.shape
```

```python
df[ (df["height"]>left1) & (df["height"]<right1) ].shape 
```

```python
df[ (df["height"]>left2) & (df["height"]<right2) ].shape 
```

```python
df[ (df["height"]>left3) & (df["height"]<right3) ].shape 
```

```python
# In std normal distribution, 68.3% data in between 1 std, 95.5% data in between 2 std, 99.7% data in between 3 std
```
